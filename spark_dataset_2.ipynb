```
spark.sql("SELECT * FROM match_table LIMIT 10").show()
```

```
spark.sql("SELECT *, CASE WHEN FTR = 'H' THEN HomeTeam WHEN FTR = 'A' THEN AwayTeam ELSE 'Tie' END AS Result FROM (SELECT * FROM match_table WHERE Div = 'D1' AND Season >= 2000 AND Season <= 2010) ").show()
```

```
#1. Who are the winners of the D1 division in the Germany Football Association (Bundesliga) in the last decade?
spark.sql("""
            WITH winner_team AS (
                    SELECT *, CASE WHEN FTR = 'H' THEN HomeTeam WHEN FTR = 'A' THEN AwayTeam ELSE 'Tie' END AS Result FROM (SELECT * FROM match_table WHERE Div = 'D1' AND Season >= 2000 AND Season <= 2010)
            )
            SELECT Result, COUNT(1) AS Total_Match_Wins FROM winner_team
            WHERE Result <> "Tie"
            GROUP BY Result
            ORDER BY Total_Match_Wins DESC LIMIT 1
          """).show()
```

```
#4. Which season of Bundesliga was the most competitive in the last decade?

spark.sql("""
            SELECT Season, COUNT(Match_ID) FROM match_table
            WHERE Season >= 2000 AND Season <= 2010
            GROUP BY Season
            ORDER BY Season
          """).show()
```

```
#5. What's the best month to watch Bundesliga?

spark.sql("""
            SELECT month, COUNT(Match_ID) AS Matches FROM match_table
            GROUP BY month
            ORDER BY Matches DESC
          """).show()
```
